sin 関数の主成分分析

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
import math
```

```python
x = np.arange(100)
print(x)
```

```python
%matplotlib inline
Y = []
for i in range(100):
    y = np.sin(2.0*math.pi*(x+i)/100.0)-0.5*np.sin(4.0*math.pi*(x+i)/100.0)
    Y.append(y)
#    plt.figure()
#    plt.plot(x,y)
#    plt.show()
#print(Y)
```

PCA of Y

```python
pca = PCA()
pca.fit(Y)
Eval = pca.explained_variance_ratio_
print('Eigen Values', Eval[:4])
plt.figure()
plt.plot(x, Eval)
plt.show()
Evec = pca.components_
for i in range(4):
    plt.figure()
    plt.plot(x, Evec[i])
    plt.show()
PCAScore=pca.transform(Y)
#print PCAScore[:,:4]
```
