Simple Plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data generation
np.random.seed(1)
x = np.arange(10)
y = np.random.rand(10)
print('x is ' + str(x))
print('y is ' + str(y))
```

```python
# plot graph
plt.plot(x, y)
plt.show()
```

f(x) = x (x - 2)(x + 2)

```python
def f(x):
    return x * (x - 2) * (x + 3)
```

```python
print(f(np.array([1, 2, 3])))
```

```python
x = np.linspace(-3, 3, 100)
print(x)
```

```python
plt.plot(x, f(x))
plt.show()
```

```python
def f2(x, w):
    return x * (x - w) * (x + 2)

x = np.linspace(-3, 3, 100)

# plot 
plt.plot(x, f2(x, 2), color='black', label='$w=2$')
plt.plot(x, f2(x, 1), color='cornflowerblue', label='$w=1$')
plt.legend(loc="upper left")
plt.ylim(-15, 15)
plt.title('$f_2(x)= x * (x - w) * (x + 2)$')
plt.xlabel('$x$')
plt.ylabel('$f_2(x)$')
plt.show()
```
