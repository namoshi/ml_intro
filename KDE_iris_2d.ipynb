ベイズ識別（アヤメのデータ（二次元））

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from matplotlib.colors import ListedColormap
```

```python
# data = np.loadtxt("iris.dat",comments='#')
iris = datasets.load_iris()
X = iris.data[:,:2]
y = iris.target
target_names = iris.target_names

print('Size of the data = ', X.shape)
```

```python
from scipy.stats import gaussian_kde
# これはエラーは出ないが誤り
kde = gaussian_kde(X.T)

# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
h = .02  # step size in the mesh
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# kdeが受け取れる形に整形
meshdata = np.vstack([xx.ravel(), yy.ravel()])
# 高さのデータ計算
z = kde.evaluate(meshdata)
print(z.shape)
print(xx.shape[0])
print(yy.shape)

# 可視化
fig = plt.figure(facecolor="w")
ax = fig.add_subplot(111, title="Kernel Density Estimation for IRIS (All data)")
ax.scatter(X[:, 0], X[:, 1], c="b")
ax.contourf(xx, yy, z.reshape(xx.shape[0], xx.shape[1]), cmap="Blues", alpha=0.5)
plt.show()
```

```python
from scipy.stats import gaussian_kde

X1 = X[(y==0),:]
X2 = X[(y==1),:]
X3 = X[(y==2),:]

kde1 = gaussian_kde(X1.T)
kde2 = gaussian_kde(X2.T)
kde3 = gaussian_kde(X3.T)

# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
h = .02  # step size in the mesh
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# kdeが受け取れる形に整形
meshdata = np.vstack([xx.ravel(), yy.ravel()])

# Probability
z1 = kde1.evaluate(meshdata)
z2 = kde2.evaluate(meshdata)
z3 = kde3.evaluate(meshdata)
print(z.shape)
print(xx.shape[0])
print(yy.shape)

# 可視化
fig = plt.figure(facecolor="w")
ax = fig.add_subplot(111, title="Kernel Density Estimation for IRIS")
ax.scatter(X1[:, 0], X1[:, 1], c="r")
ax.scatter(X2[:, 0], X2[:, 1], c="b")
ax.scatter(X3[:, 0], X3[:, 1], c="g")
ax.contourf(xx, yy, z1.reshape(xx.shape[0], xx.shape[1]), cmap="Reds", alpha=0.5)
ax.contourf(xx, yy, z2.reshape(xx.shape[0], xx.shape[1]), cmap="Blues", alpha=0.5)
ax.contourf(xx, yy, z3.reshape(xx.shape[0], xx.shape[1]), cmap="Greens", alpha=0.5)
fig.savefig("KDE_iris_2d.png")
plt.show()
```

```python
n = X.shape[0]
mdim = X.shape[1]

n1 = len(X1)
n2 = len(X2)
n3 = len(X3)

# probabilities of each class P(C_k)
P1 = 1.0 / 3.0
P2 = 1.0 / 3.0
P3 = 1.0 / 3.0
```

```python
# discriminant functions
G = np.zeros((n,3))

for i in range(n):
    dx1 = X[i,:]
    dx2 = X[i,:]
    dx3 = X[i,:]
    z1 = kde1.evaluate(dx1)
    z2 = kde2.evaluate(dx2)
    z3 = kde3.evaluate(dx3)
    G[i,0] = np.log(P1) + np.log(z1)
    G[i,1] = np.log(P2) + np.log(z2)
    G[i,2] = np.log(P3) + np.log(z3)
    
print('G\n', G)

print('Arg_max\n', np.argmax(G, axis=1))

print('Confusion Matrix\n', confusion_matrix(y, np.argmax(G, axis=1)))

x = range(n)
```

```python

plt.plot(x, G[:,0], color='red', ls='--', label=target_names[0])
plt.plot(x, G[:,1], color='blue', ls='-', label=target_names[1])
plt.plot(x, G[:,2], color='green', ls='-.', label=target_names[2])
plt.legend()
plt.title('Bayes_iris_2d')
plt.xlabel("sample no.")
plt.ylabel("discriminant function values")         
plt.savefig("bayes_iris_2d_disc.eps")
plt.show()

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

## colors
colors = ['red', 'green', 'blue']
markers = ['D', 'o', 'v']

# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
h = .02  # step size in the mesh
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z=[]
G=np.zeros((3))
for Xvec in np.c_[xx.ravel(), yy.ravel()]:
#    print(X)
    z1 = kde1.evaluate(Xvec)
    z2 = kde2.evaluate(Xvec)
    z3 = kde3.evaluate(Xvec)

    G[0] = np.log(P1) + np.log(z1)
    G[1] = np.log(P2) + np.log(z2)
    G[2] = np.log(P3) + np.log(z3)
    Z.append(np.argmax(G))
Z=np.asarray(Z)
#print('Z\n', Z)

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure()
#plt.figure(1, figsize=(4, 3))
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

# Plot also the training points
for k, c, marker, target_name  in zip([0, 1, 2], colors, markers, target_names):
    plt.scatter(X[y == k, 0], X[y == k, 1], c=c, marker=marker, label = target_name)
#plt.scatter(data[:, 0], data[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
plt.legend()
plt.title('KDE_iris_2d')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.savefig('KDE_iris_2d_class.eps')

#plt.xlim(xx.min(), xx.max())
#plt.ylim(yy.min(), yy.max())
#plt.xticks(())
#plt.yticks(())

plt.show()
```
