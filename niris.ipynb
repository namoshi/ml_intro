```python
import numpy as np

iris = np.loadtxt('iris.dat')
nr, nc = iris.shape

print(nr, nc)
```

```python
# feature vectors of each class
#
#print(iris_data)

mean_iris = np.mean(iris, axis=0)
std_iris = np.std(iris, axis=0)

niris = (iris - mean_iris)/ std_iris

print(mean_iris)
print(std_iris)
print(niris)

mean_niris = np.mean(niris, axis=0)
std_niris = np.std(niris, axis=0)

print(mean_niris)
print(std_niris)
```

```python
np.savetxt('niris.dat', niris, delimiter=' ')
```
